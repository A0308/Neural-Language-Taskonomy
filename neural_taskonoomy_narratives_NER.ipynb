```python
A = [i for i in range(0,len(sentences)-1)]
size = 10
step = 8
A = [A[i : i + size] for i in range(0, len(A), step)]
print(A)
grouped_sentences = []
for group in A:
  g = ""
  for i in group:
    g += sentences[i].strip() + '. '
  grouped_sentences.append(g)
```

```python
grouped_features = []

for text in grouped_sentences:
  group = []
  enc = tokenizer.encode(text)
  input_ids = torch.tensor(enc).unsqueeze(0)
  outputs = model(input_ids)
  enc = np.array(enc)
  ree = np.where(enc == 119)
  ree = ree[0]
  outputs = outputs[0][0].detach().numpy()
  print(outputs.shape)
  group.append(outputs[1:ree[0]+1])
  for i in range(0,len(ree)):
    if(i != len(ree)-1):
      group.append(outputs[ree[i]+1:ree[i+1]+1])
    else:
      break
      # group.append(outputs[ree[i]+1:len(outputs)-1])
  grouped_features.append(np.array(group))
```
