```python
pieman_subj_eac_L = []
pieman_subj_eac_R = []
for i in np.arange(1, 83):
    try:
        if i<10:
            files = os.listdir('sub-00'+str(i)+'/func/')
        else:
            files = os.listdir('sub-0'+str(i)+'/func/')
        num_files = sum('pieman' in s for s in files)
        temp_L = []
        temp_R = []
        if num_files==12:
            runs = ''
            if i<10:
                data_L = nib.load('sub-00'+str(i)+'/func/sub-00'+str(i)+'_task-pieman_'+runs+'space-fsaverage6_hemi-L_desc-clean.func.gii')
                data_R = nib.load('sub-00'+str(i)+'/func/sub-00'+str(i)+'_task-pieman_'+runs+'space-fsaverage6_hemi-R_desc-clean.func.gii')
            else:
                data_L = nib.load('sub-0'+str(i)+'/func/sub-0'+str(i)+'_task-pieman_'+runs+'space-fsaverage6_hemi-L_desc-clean.func.gii')
                data_R = nib.load('sub-0'+str(i)+'/func/sub-0'+str(i)+'_task-pieman_'+runs+'space-fsaverage6_hemi-R_desc-clean.func.gii')
            for j in np.arange(300):
                temp_L.append(data_L.darrays[j].data[eac_L])
                temp_R.append(data_R.darrays[j].data[eac_R])
        else:
            runs1 = 'run-1_'
            runs2 = 'run-2_'
            if i<10:
                data_L1 = nib.load('sub-00'+str(i)+'/func/sub-00'+str(i)+'_task-pieman_'+runs1+'space-fsaverage6_hemi-L_desc-clean.func.gii')
                data_L2 = nib.load('sub-00'+str(i)+'/func/sub-00'+str(i)+'_task-pieman_'+runs2+'space-fsaverage6_hemi-L_desc-clean.func.gii')
                data_R1 = nib.load('sub-00'+str(i)+'/func/sub-00'+str(i)+'_task-pieman_'+runs1+'space-fsaverage6_hemi-R_desc-clean.func.gii')
                data_R2 = nib.load('sub-00'+str(i)+'/func/sub-00'+str(i)+'_task-pieman_'+runs2+'space-fsaverage6_hemi-R_desc-clean.func.gii')
            else:
                data_L1 = nib.load('sub-0'+str(i)+'/func/sub-0'+str(i)+'_task-pieman_'+runs1+'space-fsaverage6_hemi-L_desc-clean.func.gii')
                data_L2 = nib.load('sub-0'+str(i)+'/func/sub-0'+str(i)+'_task-pieman_'+runs2+'space-fsaverage6_hemi-L_desc-clean.func.gii')
                data_R1 = nib.load('sub-0'+str(i)+'/func/sub-0'+str(i)+'_task-pieman_'+runs1+'space-fsaverage6_hemi-R_desc-clean.func.gii')
                data_R2 = nib.load('sub-0'+str(i)+'/func/sub-0'+str(i)+'_task-pieman_'+runs2+'space-fsaverage6_hemi-R_desc-clean.func.gii')
            for j in np.arange(300):
                temp_L.append(np.mean([data_L1.darrays[j].data[eac_L], data_L2.darrays[j].data[eac_L]], axis=0))
                temp_R.append(np.mean([data_R1.darrays[j].data[eac_R], data_R2.darrays[j].data[eac_R]], axis=0))
        pieman_subj_eac_L.append(np.array(temp_L))
        pieman_subj_eac_R.append(np.array(temp_R))
    except:
        print(i)
        continue
```
